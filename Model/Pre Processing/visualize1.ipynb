```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('semvii_studentdata.csv')
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='age', data=df)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['study_hour'], bins=20, kde=True)
plt.title('Study Hours Distribution')
plt.xlabel('Study Hours')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['total_marks'], bins=20, kde=True)
plt.title('Total Marks Distribution')
plt.xlabel('Total Marks')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['CGPA'], bins=20, kde=True)
plt.title('CGPA Distribution')
plt.xlabel('CGPA')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df[['age', 'study_hour', 'total_marks', 'percentage', 'CGPA']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
